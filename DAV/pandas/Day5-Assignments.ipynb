The zip() function pairs corresponding elements from both lists, and the resulting tuples are used to create the DataFrame. The column names for the DataFrame are set as “RID” and “RDate”.

- res_df['RDate'] = pd.to_datetime(res_df['RDate']) : This line converts the “RDate” column of the res_df DataFrame to pandas DateTime format using the pd.to_datetime() function. This allows for easier manipulation and extraction of date-related information.

- res_df['RMonth'] = res_df['RDate'].dt.month : This line creates a new column called “RMonth” in the res_df DataFrame, which extracts the month component from the “RDate” column using the .dt.month attribute. The month values are represented as integers.

- res_df['RYear'] = res_df['RDate'].dt.year : This line creates another new column called “RYear” in the res_df DataFrame, which extracts the year component from the “RDate” column using the .dt.year attribute. The year values are represented as integers.

- res_df['RDay'] = res_df['RDate'].dt.day : This line creates a new column called “RDay” in the res_df DataFrame, which extracts the day component from the “RDate” column using the .dt.day attribute. The day values are represented as integers.

Finally, the modified res_df DataFrame is returned as the output of the function.

```python

import pandas as pd
def solve(reg_id, reg_dates):
    res_df = pd.DataFrame(zip(reg_id, reg_dates), columns=["RID", "RDate"])
    """
    input:
    reg_id -> id of the input dataframe
    reg_dates -> dates of the input dataframe
    res_df -> the input dataframe 
    
    output:
    return the resultant dataframe after performing the required operation
    """
    # YOUR CODE GOES HERE
    res_df = pd.DataFrame(zip(reg_id, reg_dates), columns=["RID", "RDate"])
    res_df['RDate'] = pd.to_datetime(res_df['RDate'])
    res_df['RMonth'] = res_df['RDate'].dt.month
    res_df['RYear'] = res_df['RDate'].dt.year
    res_df['RDay'] = res_df['RDate'].dt.day 

    return res_df

solve(reg_id,reg_dates)
```

Write a python program to do so using apply function.

```python
import pandas as pd

data = {
    'name':["Ram","Shyam","Mukesh","Suresh"],
    'age':[10,18,60,50]
}

a = pd.DataFrame(data)
a
```

```python
import pandas as pd
def replace_ages(a):
    '''a is a dataframe with columns ['name', 'age']
       Output A dataframe with same columns but with changed records as per required is expected to be returned'''

    a['age'] = a['age'].apply(lambda x: 'kid' if x < 18 else ('adult' if x >= 18 and x <=50 else 'senior'))

    return a

replace_ages(a)
```

Note: Recall how we group and merge data in Pandas.

Use the code below in order to create the given dataframe:

            'subject': ['NN', 'DL', 'ML', 'Prob', 'DL', 'ML', 'DL', 'NN', 'NN', 'Prob', 'Prob', 'ML'],

```python
import pandas as pd

data = {
    'roll_no': [1, 2, 1, 3, 1, 3, 3, 3, 2, 2, 1, 2],
    'subject': ['NN', 'DL', 'ML', 'Prob', 'DL', 'ML', 'DL', 'NN', 'NN', 'Prob', 'Prob', 'ML'],
    'marks': [97, 63, 63, 71, 64, 90, 66, 46, 74, 62, 94, 67]
}

df = pd.DataFrame(data)
df
```

```python
import pandas as pd

def get_marks(df):
    '''
    INPUT: df -> dataframe
    
    OUTPUT: result -> dataframe
    '''
    
    ### STEP 1: group the df using subject and get max marks
    
    max_marks = df.groupby('subject')['marks'].max()
    
    ### STEP 2: reset the index
    
    max_marks = max_marks.reset_index()
    
    
    ### STEP 3: merge with original df to get roll id
    
    result = pd.merge(max_marks,df,on=('subject','marks'),how='inner')
    
    return result

get_marks(df)
```
