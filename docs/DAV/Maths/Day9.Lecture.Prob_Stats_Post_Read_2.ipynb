## 📈 **Log-Normal Distribution**

**Definition:**  
A **<font color="green">Log-Normal Distribution</font>** describes a random variable whose logarithm is normally distributed. Unlike the **Normal Distribution**, it is **skewed** and only takes positive values.

**Real-World Example:**  
- **Stock Prices:** Often modeled using log-normal distributions because they cannot be negative and exhibit right skewness.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu, sigma = 0, 0.25  # mean and standard deviation of the underlying normal distribution
size = 1000

# Generate log-normal data
log_normal_data = np.random.lognormal(mean=mu, sigma=sigma, size=size)

# Plot histogram
plt.hist(log_normal_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Log-Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```


**Interpretation:**  
The histogram shows a right-skewed distribution, characteristic of **<font color="green">Log-Normal</font>** data.

## 🔢 **Poisson Distribution**

**Definition:**  
The **<font color="green">Poisson Distribution</font>** models the number of times an event occurs in a fixed interval of time or space. It is defined by the parameter **λ (lambda)**, representing the average rate of occurrence.

**Use Cases:**  
- **Count of Events:** Such as the number of emails received per hour or the number of accidents at an intersection.

```python
# Parameter
lambda_poisson = 3
size = 1000

# Generate Poisson data
poisson_data = np.random.poisson(lam=lambda_poisson, size=size)

# Plot histogram
plt.hist(poisson_data, bins=range(0, max(poisson_data)+1), color='lightgreen', edgecolor='black', align='left')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')
plt.show()
```

**Interpretation:**  
The histogram displays the probability of different event counts, centered around **λ = 3**.

## ⏱️ **Exponential Distribution**

**Definition:**  
The **<font color="green">Exponential Distribution</font>** models the time between consecutive events in a Poisson process. It is characterized by the rate parameter **λ (lambda)**.

**Perspective:**  
- **Time Between Events:** Such as the time between arrivals of customers at a store.



```python
# Parameter
lambda_exp = 1.5  # rate parameter
size = 1000

# Generate Exponential data
exponential_data = np.random.exponential(scale=1/lambda_exp, size=size)

# Plot histogram
plt.hist(exponential_data, bins=30, color='salmon', edgecolor='black')
plt.title('Exponential Distribution')
plt.xlabel('Time Between Events')
plt.ylabel('Frequency')
plt.show()
```


**Interpretation:**  
The histogram shows a right-skewed distribution, indicating shorter times between events are more common.

---

## 🎲 **Geometric Distribution**

**Definition:**  
The **<font color="green">Geometric Distribution</font>** models the number of trials needed to achieve the first success in a sequence of independent Bernoulli trials.

**Use Case:**  
- **First Success:** Such as the number of coin flips until the first heads appears.

```python
# Parameter
p = 0.2  # probability of success
size = 1000

# Generate Geometric data
geometric_data = np.random.geometric(p=p, size=size)

# Plot histogram
plt.hist(geometric_data, bins=range(1, max(geometric_data)+1), color='violet', edgecolor='black', align='left')
plt.title('Geometric Distribution')
plt.xlabel('Number of Trials Until First Success')
plt.ylabel('Frequency')
plt.show()
```


**Interpretation:**  
The histogram peaks at lower trial numbers, decreasing as the number of trials increases.

## 🔄 **Box-Cox Transformation**

**Why Transform Data?** 💡  
Data transformations can **correct skewness** and make the data **more normal**, which is beneficial for various statistical analyses and modeling techniques.

**<font color="skyblue">Box-Cox Transformation</font>:**  
A family of power transformations that aim to stabilize variance and make the data more normal distribution-like.

```python
from scipy import stats

# Generate skewed data
np.random.seed(0)
skewed_data = np.random.exponential(scale=2, size=1000)

# Apply Box-Cox Transformation
transformed_data, lambda_bc = stats.boxcox(skewed_data)

# Plot before and after
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data
axes[0].hist(skewed_data, bins=30, color='orange', edgecolor='black')
axes[0].set_title('Original Skewed Data')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Transformed data
axes[1].hist(transformed_data, bins=30, color='lightblue', edgecolor='black')
axes[1].set_title('Box-Cox Transformed Data')
axes[1].set_xlabel('Transformed Value')
axes[1].set_ylabel('Frequency')

plt.show()

print(f'Optimal Box-Cox lambda: {lambda_bc:.2f}')
```


**Interpretation:**  
The left histogram shows the original **skewed** data, while the right histogram illustrates the **transformed** data, which is more symmetric and closer to a normal distribution. The optimal lambda value is also displayed.
