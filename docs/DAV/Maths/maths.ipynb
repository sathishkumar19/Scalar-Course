````python
import scipy.stats as stats

# Define the discrete uniform distribution for digits 0 through 9
# The probability for each digit is 1/10 since they are equally likely
dist = stats.randint(low=0, high=10)
# Probability that the last digit is 7

# Probability that the last digit is less than 4
prob_last_digit_7 = dist.pmf(7)
print(f"Probability that the last digit is 7: {prob_last_digit_7}")

prob_last_digit_less_than_4 = dist.cdf(3)
print(f"Probability that the last digit is less than 4: {prob_last_digit_less_than_4}")
````

````python

# Suppose you reside in a city with a large population. The weight of people living in your city follows a Gaussian distribution with a mean of 60 kgs with a standard deviation of 10 kgs.
# You weigh 65 kg. What is the probability that a randomly selected person in your city is heavier than you?
# Round off the answer to 4 decimal points)

import scipy.stats as stats

# Given data
mean = 60
std_dev = 10
my_weight = 65

# Calculate the z-score
z_score = (my_weight - mean) / std_dev
z_score

# Calculate the probability that a randomly selected person in your city is heavier than you
prob_heavier = stats.norm.sf(z_score)
round(prob_heavier, 4)
````

## Probability Mass Function (PMF)

A Probability Mass Function (PMF) gives the probability that a discrete random variable is exactly equal to some value. The PMF is defined for discrete random variables and can be represented as:

\[ P(X = x) = f(x) \]

where \( P(X = x) \) is the probability that the random variable \( X \) takes the value \( x \).

### Example Python Code

Here is an example of how to calculate the PMF using Python with the `scipy.stats` module:

```python
import scipy.stats as stats

# Define the discrete uniform distribution for digits 0 through 9
dist = stats.randint(low=0, high=10)

# Calculate the PMF for a specific value, e.g., 7
pmf_value = dist.pmf(7)
print(f"PMF for value 7: {pmf_value}")

# Calculate the PMF for all possible values in the distribution
pmf_values = [dist.pmf(x) for x in range(10)]
print(f"PMF values for all possible outcomes: {pmf_values}")

````python
# An unknown distribution has a mean of 90 and a standard deviation of 15. Samples of size n = 35 are drawn randomly from the population.
# Find the probability that the sample mean is between 85 and 92.

import scipy.stats as stats # import the stats module from the scipy library
from scipy.stats import norm # normal distribution

# Given data
mean = 90
std_dev = 15
sample_size = 35

# find the probability that the sample mean is between 85 and 92
z_score_85 = (85 - mean) / (std_dev / (sample_size ** 0.5))
z_score_92 = (92 - mean) / (std_dev / (sample_size ** 0.5))

prob_85_to_92 = norm.cdf(z_score_92) - norm.cdf(z_score_85)
prob_85_to_92
````

````python
# You go to a party of 500 guests. What is the probability that exactly one other guest has the same birthday as you?
# Note: Exclude birthdays on February 29.

import scipy.stats as stats # import the stats module from the scipy library
from scipy.stats import binom # binomial distribution

# Given data
n = 500
p = 1 / 365

# find the probability that exactly one other guest has the same birthday as you

prob_one_other_guest_same_birthday = binom.pmf(1, n, p)
prob_one_other_guest_same_birthday
````

````python
# The tip percentage for a delivery person has a mean value of 28% and a standard deviation of 7%.
# What is the probability that the sample mean tip percentage for a random sample of 50 bills is between 26% and 31%?

import scipy.stats as stats # import the stats module from the scipy library
from scipy.stats import norm # normal distribution

# Given data
mean = 28
std_dev = 7
sample_size = 50

# find the probability that the sample mean tip percentage is between 26% and 31%
z_score_26 = (26 - mean) / (std_dev / (sample_size ** 0.5))
z_score_31 = (31 - mean) / (std_dev / (sample_size ** 0.5))

prob_26_to_31 = norm.cdf(z_score_31) - norm.cdf(z_score_26)
prob_26_to_31
````

````python
#An observer counts 240 vehicles per hour (3600 seconds) at a specific vehicle arrival location on a highway.
#Assume that the arrivals follow the poisson distribution, What is the probability of one vehicle arriving over a 30-second time interval?

import scipy.stats as stats # import the stats module from the scipy library
from scipy.stats import poisson # po

# Given data
arrival_rate_per_hour = 240
arrival_rate_per_second = arrival_rate_per_hour / 3600
time_interval = 30  # seconds   

# find the probability of one vehicle arriving over a 30-second time interval
prob_one_vehicle = poisson.pmf(1, arrival_rate_per_second * time_interval)
prob_one_vehicle
````

````python
# You are working for the Department of Endangered Species. Along with your team, you have bred a certain species of endangered fish in your pond.
# On taking a sample of 100 fishes, the calculated average length of this species came out to be 24cm. It is known that the population standard deviation in the length of these fishes is 8cm.
# Estimate the mean length confidence interval of the fishes in the pond with a confidence of 95 percent.
# Note: z-multiplier for 95% is 1.96.


import scipy.stats as stats # import the stats module from the scipy library
from scipy.stats import norm # normal


# Given data
fish_sample_size = 100
avg_length = 24
population_std_dev = 8

# Calculate the standard error of the mean
sem_value = population_std_dev / (fish_sample_size ** 0.5)

# Calculate the margin of error
z_multiplier = 1.96
margin_of_error = z_multiplier * sem_value

# Calculate the confidence interval
confidence_interval = (avg_length - margin_of_error, avg_length + margin_of_error)
confidence_interval

````

````python
# Q2. Power failure
# If electricity power failures occur according to a poisson distribution with an average of three failures every 20 weeks.
# Calculate the probability that there will not be more than one power failure during a week.


import scipy.stats as stats # import the stats module from the scipy library
from scipy.stats import poisson # po

# Given data
average_failures_per_week = 3 / 20
time_interval = 1  # week

# Calculate the probability that there will not be more than one power failure during a week
powerfailure_per_week = poisson.cdf(1, average_failures_per_week * time_interval)
powerfailure_per_week



````
