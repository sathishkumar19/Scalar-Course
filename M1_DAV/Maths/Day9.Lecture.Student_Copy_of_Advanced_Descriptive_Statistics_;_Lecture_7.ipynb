## 📈 2. Descriptive Statistics

Descriptive statistics provide **simple summaries** about the **sample** and the **measures**. They form the foundation for virtually every quantitative analysis of data.




### 🏹 **Measures of Central Tendency**

Central tendency measures describe the **center** of a data distribution. The main measures are **<font color="magenta">Mean</font>**, **<font color="skyblue">Median</font>**, and **<font color="blue">Mode</font>**.



#### 🔹 **<font color="greed">1. Mean</font>**

The **<font color="greed">Mean</font>** is the **average value**.

**Formula:**

$$
\text{Mean} (\mu) = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

```python
# Calculate Mean Runs
mean_runs = df['runs'].mean()
print(f"Mean Runs: {mean_runs:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Histogram with Mean
plt.figure(figsize=(8,6))
sns.histplot(df['runs'], bins=30, color='magenta', kde=True)
plt.axvline(df['runs'].mean(), color='red', linestyle='dashed', linewidth=2, label=f"Mean: {df['runs'].mean():.2f}")
plt.title('Distribution of Runs with Mean')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```


**Explanation:**  
The mean gives us the **average runs** Sachin scored per match.


#### 🔹 **<font color="skyblue">2. Median</font>**

The **<font color="skyblue">Median</font>** is the **middle value** when data is ordered.

**Formula:**

For an ordered dataset with $ N $ observations:
- If $ N $ is odd:
  $$
  \text{Median} = x_{\left(\frac{N+1}{2}\right)}
  $$
- If $ N $ is even:
  $$
  \text{Median} = \frac{x_{\left(\frac{N}{2}\right)} + x_{\left(\frac{N}{2} + 1\right)}}{2}
  $$

```python
# Calculate Median Runs
median_runs = df['runs'].median()
print(f"Median Runs: {median_runs}")
```

```python
# Plot Histogram with Median
plt.figure(figsize=(8,6))
sns.histplot(df['runs'], bins=30, color='skyblue', kde=True)
plt.axvline(df['runs'].median(), color='green', linestyle='dashed', linewidth=2, label=f"Median: {df['runs'].median()}")
plt.title('Distribution of Runs with Median')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Explanation:**  
The median provides the **central value** of Sachin's runs, minimizing the effect of outliers.

#### 🔹 **<font color="skyblue">3. Mode</font>**

The **<font color="skyblue">Mode</font>** is the **most frequently occurring value**.

**Formula:**

There is no mathematical formula for the mode, as it is determined by the frequency of values:

$$
\text{Mode} = \text{Value with the highest frequency}
$$

```python
# Calculate Mode Runs
mode_runs = df['runs'].mode()[0]
print(f"Mode Runs: {mode_runs}")
```

```python
# Plot Histogram with Mode
plt.figure(figsize=(8,6))
sns.histplot(df['runs'], bins=30, color='blue', kde=True)
mode_runs = df['runs'].mode()[0]
plt.axvline(mode_runs, color='orange', linestyle='dashed', linewidth=2, label=f"Mode: {mode_runs}")
plt.title('Distribution of Runs with Mode')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Explanation:**  
The mode indicates the **most common score** Sachin achieved in his matches.

```python
# Combined Plot for Mean, Median, and Mode
plt.figure(figsize=(10,7))
sns.histplot(df['runs'], bins=30, color='lightgray', kde=True)

# Calculate statistics
mean = df['runs'].mean()
median = df['runs'].median()
mode = df['runs'].mode()[0]

# Plot lines
plt.axvline(mean, color='magenta', linestyle='dashed', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(median, color='skyblue', linestyle='dashed', linewidth=2, label=f"Median: {median}")
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=2, label=f"Mode: {mode}")

plt.title('Distribution of Runs with Mean, Median, and Mode')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### 🏹 **Measures of Variability**

Variability measures describe the **spread** of the data. The main measures are **<font color="green">Range</font>**, **<font color="orange">Variance</font>**, and **<font color="greed">Standard Deviation</font>**.

#### 🔹 **<font color="green">1. Range</font>**

The **<font color="green">Range</font>** is the **difference between the maximum and minimum values**.

**Formula:**

$$
\text{Range} = \text{Max}(x) - \text{Min}(x)
$$


```python
# Calculate Range of Runs
range_runs = df['runs'].max() - df['runs'].min()

print(f"({df['runs'].min()},{df['runs'].max()})")
print(f"Range of Runs: {range_runs}")
```

```python
# Plot Histogram with Range
plt.figure(figsize=(8,6))
sns.histplot(df['runs'], bins=30, color='purple', kde=True)
min_run = df['runs'].min()
max_run = df['runs'].max()
plt.axvline(min_run, color='red', linestyle='dashed', linewidth=2, label=f"Min: {min_run}")
plt.axvline(max_run, color='green', linestyle='dashed', linewidth=2, label=f"Max: {max_run}")
plt.title('Distribution of Runs with Range')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### 🔹 **<font color="orange">2. Variance</font>**

The **<font color="orange">Variance</font>** measures the **average degree to which each point differs from the mean**.

**Formula:**

For a sample:
$$
\text{Variance} (s^2) = \frac{1}{N-1} \sum_{i=1}^{N} (x_i - \mu)^2
$$

```python
# Calculate Variance of Runs
variance_runs = df['runs'].var()
print(f"Variance of Runs: {variance_runs:.2f}")
```

**Explanation:**  
Variance provides insight into the **consistency** of Sachin's performances.

#### 🔹 **<font color="greed">3. Standard Deviation</font>**

The **<font color="greed">Standard Deviation</font>** is the **square root of variance**, indicating data dispersion.

**Formula:**

$$
\text{Standard Deviation} (\sigma) = \sqrt{\text{Variance}} = \sqrt{\frac{1}{N-1} \sum_{i=1}^{N} (x_i - \mu)^2}
$$


```python
# Calculate Standard Deviation of Runs
std_dev_runs = df['runs'].std()
print(f"Standard Deviation of Runs: {std_dev_runs:.2f}")
```

```python
import numpy as np

# Plot Histogram with Standard Deviation
plt.figure(figsize=(8,6))
sns.histplot(df['runs'], bins=30, color='teal', kde=True)
mean = df['runs'].mean()
std_dev = df['runs'].std()
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=2, label=f"Mean + 1 SD: {mean + std_dev:.2f}")
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=2, label=f"Mean - 1 SD: {mean - std_dev:.2f}")
plt.title('Distribution of Runs with Standard Deviation')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Explanation:**  
Standard deviation offers a **more interpretable** measure of variability in Sachin's scores.

```python
# Plot Histogram with Multiple Standard Deviations
plt.figure(figsize=(10,7))
sns.histplot(df['runs'], bins=30, color='lightgreen', kde=True)

# Calculate statistics
mean = df['runs'].mean()
std_dev = df['runs'].std()

# Plot mean and standard deviations
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean:.2f}")
plt.axvline(mean + std_dev, color='orange', linestyle='dashed', linewidth=2, label=f"Mean + 1 SD")
plt.axvline(mean - std_dev, color='orange', linestyle='dashed', linewidth=2, label=f"Mean - 1 SD")
plt.axvline(mean + 2*std_dev, color='purple', linestyle='dashed', linewidth=2, label=f"Mean + 2 SD")
plt.axvline(mean - 2*std_dev, color='purple', linestyle='dashed', linewidth=2, label=f"Mean - 2 SD")

plt.title('Distribution of Runs with Multiple Standard Deviations')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## 🎲 3. Random Variables

A **<font color="magenta">Random Variable</font>** is a variable whose possible values are numerical outcomes of a **random phenomenon**. Random variables are fundamental in statistics and probability as they provide a way to quantify outcomes and analyze data distributions.

### 🔹 **Discrete vs. Continuous Random Variables**

Understanding the types of random variables is crucial for selecting appropriate statistical methods and distributions.

#### **<font color="skyblue">Discrete Random Variables</font>**

- **Definition:**  
  **Discrete Random Variables** take on a **countable number of distinct values**. These values are typically integers and can be listed individually.

- **Examples:**  
  - **Number of Runs:** The total runs Sachin scores in a match (e.g., 0, 50, 100).
  - **Number of Centuries:** Count of matches where Sachin scores a century.

#### **<font color="greed">Continuous Random Variables</font>**

- **Definition:**  
  **Continuous Random Variables** take on an **infinite number of possible values** within a given range. These values are often measured and can include decimals.

- **Examples:**  
  - **Strike Rate (`sr`):** Sachin's strike rate in a match (e.g., 75.50, 82.30).
  - **Balls Faced (`bf`):** Number of balls Sachin faces, which can vary continuously over time.

**💡 Note:**  
Identifying whether a random variable is discrete or continuous helps in determining the appropriate probability distribution and statistical analysis techniques to apply.

### 📊 **Probability Functions**

Random variables are characterized by their probability functions, which describe how probabilities are distributed over the possible values of the variable. These functions are essential for understanding the behavior and likelihood of different outcomes.



#### 🔹 **1. Probability Mass Function (PMF)**

- **Definition:**  
  The **<font color="orange">Probability Mass Function (PMF)</font>** applies to **discrete random variables**. It provides the probability that a discrete random variable is exactly equal to some value.

- **Formula:**
  
  $$
  P(X = x) = p(x)
  $$

  Where:
  - $ P(X = x) $ is the probability that the random variable $ X $ equals $ x $.
  - $ p(x) $ is the probability mass at value $ x $.

- **Examples:**  
  1. **Number of Centuries:**  
     The probability that Sachin scores exactly 1 century in a match.
  
  2. **Number of Fours:**  
     The probability that Sachin hits exactly 4 fours in a match.

- **💡 Explanation:**  
  The PMF assigns a probability to each possible discrete outcome. For instance, if Sachin scores 50 runs in a match, the PMF tells us the probability of this exact event occurring.


```python
# Probability Mass Function (PMF) Examples

# Example 1: Probability of scoring exactly 1 century
from scipy.stats import binom

# Assuming the probability of scoring a century in a match is p
p_century = df['century'].mean()

# PMF for scoring exactly 1 century in a single match
pmf_century_1 = binom.pmf(k=1, n=1, p=p_century)
print(f"P(Scoring exactly 1 century in 1 match) = {pmf_century_1:.4f}")
```

```python
# Example 2: Probability of hitting exactly 4 fours in a match
# First, identify the distribution (assuming it's Poisson for illustration)
from scipy.stats import poisson

# Estimate lambda as the average number of fours
lambda_fours = df['fours'].mean()

# PMF for hitting exactly 4 fours
pmf_fours_4 = poisson.pmf(k=4, mu=lambda_fours)
print(f"P(Hitting exactly 4 fours in a match) = {pmf_fours_4:.4f}")
```

#### 🔹 **2. Probability Density Function (PDF)**

- **Definition:**  
  The **<font color="green">Probability Density Function (PDF)</font>** applies to **continuous random variables**. It describes the relative likelihood for the random variable to take on a given value.

- **Formula:**
  
  $$
  f(x) = \frac{d}{dx}F(x)
  $$

  Where:
  - $ f(x) $ is the probability density at value $ x $.
  - $ F(x) $ is the cumulative distribution function at value $ x $.

- **Examples:**  
  1. **Strike Rate (`sr`):**  
     Modeling the distribution of Sachin's strike rates across all matches.
  
  2. **Balls Faced (`bf`):**  
     Modeling the distribution of the number of balls Sachin faces in matches.

- **💡 Explanation:**  

  The PDF provides a continuous curve representing the density of probabilities across a range of values. Unlike PMF, the PDF does not give probabilities for exact values but rather the likelihood of the variable falling within a specific interval.

- **Visualization:**  
  A **<font color="green">Histogram</font>** of Sachin's strike rates can approximate the PDF, showing the frequency of strike rates within specified intervals.


```python
# Probability Density Function (PDF) Examples

# Example 1: Strike Rate (`sr`) Distribution
from scipy.stats import norm

# Parameters for Normal Distribution
mu_sr = df['sr'].mean()
sigma_sr = df['sr'].std()

# Generate values for PDF
x_sr = np.linspace(mu_sr - 4*sigma_sr, mu_sr + 4*sigma_sr, 1000)
pdf_sr = norm.pdf(x_sr, mu_sr, sigma_sr)

# Plotting
plt.figure(figsize=(8,6))
sns.histplot(df['sr'], bins=30, kde=False, color='skyblue', stat='density', label='Strike Rate Histogram')
plt.plot(x_sr, pdf_sr, color='green', label='Normal PDF')
plt.title('PDF of Sachin’s Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Example 2: Balls Faced (`bf`) Distribution
# Assuming Balls Faced follows a Normal Distribution
mu_bf = df['bf'].mean()
sigma_bf = df['bf'].std()

# Generate values for PDF
x_bf = np.linspace(mu_bf - 4*sigma_bf, mu_bf + 4*sigma_bf, 1000)
pdf_bf = norm.pdf(x_bf, mu_bf, sigma_bf)

# Plotting
plt.figure(figsize=(8,6))
sns.histplot(df['bf'], bins=30, kde=False, color='orange', stat='density', label='Balls Faced Histogram')
plt.plot(x_bf, pdf_bf, color='purple', label='Normal PDF')
plt.title('PDF of Sachin’s Balls Faced')
plt.xlabel('Balls Faced')
plt.ylabel('Density')
plt.legend()
plt.show()
```

#### 🔹 **3. Cumulative Distribution Function (CDF)**

- **Definition:**  
  The **<font color="orange">Cumulative Distribution Function (CDF)</font>** applies to both **discrete and continuous random variables**. It represents the probability that the random variable $ X $ will take a value **less than or equal to** $ x $.

- **Formula:**
  
  $$
  F(x) = P(X \leq x) =
  \begin{cases}
    \sum_{t \leq x} p(t) & \text{(Discrete)} \\
    \int_{-\infty}^{x} f(t) \, dt & \text{(Continuous)}
  \end{cases}
  $$

  Where:
  - $ F(x) $ is the CDF at value $ x $.
  - $ p(t) $ is the PMF for discrete variables.
  - $ f(t) $ is the PDF for continuous variables.

- **Examples:**  
  1. **Number of Centuries:**  
     The probability that Sachin scores **up to** 2 centuries in a match.
  
  2. **Strike Rate (`sr`):**  
     The probability that Sachin's strike rate is **less than or equal to** 80 in a match.

- **💡 Explanation:**  
  The CDF accumulates probabilities from the lowest value up to a specified value $ x $. It provides a comprehensive view of the distribution, showing the likelihood of the variable falling within a range.

- **Visualization:**  
  The **<font color="orange">CDF Plot</font>** accumulates probabilities from the left up to a given value $ x $, providing a comprehensive view of the distribution.

```python
# Example 1: CDF for Strike Rate (`sr`)
# Using Normal Distribution
from scipy.stats import norm

# Parameters
mu_sr = df['sr'].mean()
sigma_sr = df['sr'].std()

# Calculate CDF for strike rate < 80
x_sr = 80
cdf_less_80 = norm.cdf(80, mu_sr, sigma_sr)
print(f"P(Strike Rate < 80) = {cdf_less_80:.4f}")
```

```python
# Example 2: Probability of Strike Rate between 80 and 100
cdf_80 = norm.cdf(80, mu_sr, sigma_sr)
cdf_100 = norm.cdf(100, mu_sr, sigma_sr)

prob_80_100 = cdf_100 - cdf_80
print(f"P(80 ≤ Strike Rate ≤ 100) = {prob_80_100:.4f}")
```

```python
# Example 3: Probability of Strike Rate > 120
cdf_120 = norm.cdf(120, mu_sr, sigma_sr)
prob_greater_120 = 1 - cdf_120
print(f"P(Strike Rate > 120) = {prob_greater_120:.4f}")
```

## 🎲 4. Bernoulli & Binomial Distributions

Understanding probability distributions is essential for modeling and analyzing different types of data. In this section, we'll explore the **<font color="orange">Bernoulli Distribution</font>** and the **<font color="green">Binomial Distribution</font>**, their key properties, and real-life applications using **Sachin Tendulkar’s ODI Career** dataset.

### 🔹 **Bernoulli Distribution**

- **Definition:**  
  The **<font color="orange">Bernoulli Distribution</font>** is a **discrete probability distribution** applicable to a single trial that has only two possible outcomes: **success** (1) or **failure** (0).

- **Parameters:**  
  - $ p $: Probability of success (where $ 0 \leq p \leq 1 $)

- **Formula:**
  
  $$
  P(X = x) =
  \begin{cases}
    p & \text{if } x = 1 \\
    1 - p & \text{if } x = 0
  \end{cases}
  $$

- **Example:**  
  **Scoring a Century:**  
  Consider each match as a Bernoulli trial where:
  - **Success (1):** Sachin scores a century.
  - **Failure (0):** Sachin does not score a century.
  
  The probability of success $ p $ can be estimated as the proportion of matches where Sachin scored a century.

- **💡 Real-Life Application:**  
  Determining the probability of a specific event occurring in a single trial, such as winning a match, hitting a boundary, etc.



### 🔹 **Binomial Distribution**

- **Definition:**  
  The **<font color="green">Binomial Distribution</font>** is a **discrete probability distribution** that models the number of successes in a fixed number of independent Bernoulli trials.

- **Parameters:**  
  - $ n $: Number of trials.
  - $ p $: Probability of success on a single trial.

- **Formula:**
  
  $$
  P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
  $$

  Where:
  - $ \binom{n}{k} $ is the binomial coefficient.
  - $ k $: Number of successes.

- **Example:**  
  **Number of Centuries in Multiple Matches:**  
  If we consider $ n = 10 $ matches, the Binomial distribution can model the probability of Sachin scoring exactly $ k $ centuries in those 10 matches.

- **💡 Real-Life Application:**  
  Calculating the likelihood of a specific number of successes over multiple trials, such as the number of matches won, number of high scores, etc.

### 📊 **Probability Functions**

Probability distributions are characterized by their probability functions, which describe how probabilities are assigned to different outcomes. Below, we'll delve into the **PMF** of Bernoulli and Binomial distributions, along with practical examples using Sachin's ODI data.

#### 🔹 **Probability Mass Function (PMF)**

- **Bernoulli PMF:**
  
  $$
  P(X = x) =
  \begin{cases}
    p & \text{if } x = 1 \\
    1 - p & \text{if } x = 0
  \end{cases}
  $$

- **Binomial PMF:**
  
  $$
  P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
  $$

**Examples and Code Demonstrations**

Below are practical examples demonstrating how to calculate and visualize the **Bernoulli** and **Binomial** PMFs using Sachin's ODI data.

```python
# Probability Mass Function (PMF) for Bernoulli Distribution

from scipy.stats import bernoulli

# Probability of scoring a century in a single match
p_century = df['century'].mean()
print(f"Probability of scoring a century (p): {p_century:.4f}")

# Bernoulli PMF for x = 0 (No Century) and x = 1 (Century)
x = [0, 1]
pmf = bernoulli.pmf(x, p_century)
for outcome, probability in zip(x, pmf):
    status = "Century" if outcome == 1 else "No Century"
    print(f"P({status}) = {probability:.4f}")
```

**Example 2: Binomial Distribution - Probability of Scoring Exactly 3 Centuries in 5 Matches**

```python
# Probability Mass Function (PMF) for Binomial Distribution

from scipy.stats import binom

# Parameters
n_matches = 5  # Number of matches
k_centuries = 3  # Desired number of centuries
p_century = df['century'].mean()

# Binomial PMF
pmf_binom = binom.pmf(k_centuries, n_matches, p_century)
print(f"P(Scoring exactly {k_centuries} centuries in {n_matches} matches) = {pmf_binom:.4f}")
```

**Example 3: Binomial Distribution - Probability of Scoring at Least 2 Centuries in 4 Matches**

```python
# Cumulative Distribution Function (CDF) for Binomial Distribution

from scipy.stats import binom

# Parameters
n_matches = 4
k_min = 2
p_century = df['century'].mean()

# Probability of scoring up to (k_min - 1) centuries
cdf_up_to = binom.cdf(k_min - 1, n_matches, p_century)

# Probability of scoring at least k_min centuries
prob_at_least = 1 - cdf_up_to
print(f"P(Scoring at least {k_min} centuries in {n_matches} matches) = {prob_at_least:.4f}")
```

### 📊 **Visualizing Bernoulli & Binomial Distributions**

Visual representations help in comprehending the probability distributions and their implications.

```python
# Bernoulli Distribution Visualization

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli

# Parameters
p = df['century'].mean()
x = [0, 1]
pmf = bernoulli.pmf(x, p)

# Plotting
plt.figure(figsize=(6,4))
sns.barplot(x=x, y=pmf, palette='viridis')
plt.xticks([0, 1], ['No Century', 'Century'])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution of Scoring a Century')
plt.ylim(0, 1)
plt.show()
```

```python
# Binomial Distribution Visualization

from scipy.stats import binom

# Parameters
n = 10  # Number of matches
p = df['century'].mean()
k = range(0, n+1)

# Binomial PMF
pmf_binom = binom.pmf(k, n, p)

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(x=k, y=pmf_binom, palette='coolwarm')
plt.xlabel('Number of Centuries')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p:.2f})')
plt.show()
```

## 🌟 4. Normal (Gaussian) Distribution

The **<font color="purple">Normal (Gaussian) Distribution</font>** is one of the most fundamental and widely used probability distributions in statistics and machine learning. It is characterized by its symmetric, bell-shaped curve and plays a crucial role in various statistical analyses and inference methods.

### 🔹 **Definition**

- **Normal Distribution:**  
  The Normal Distribution is a **continuous probability distribution** that is symmetric about its mean, depicting that data near the mean are more frequent in occurrence than data far from the mean. In graph form, the Normal Distribution appears as a bell-shaped curve.

### 🔹 **Parameters**

The Normal Distribution is defined by two parameters:

1. **Mean ($ \mu $)**
   - **Description:** Represents the central tendency or the "average" of the distribution.
   - **Interpretation:** Indicates the location of the peak of the bell curve.

2. **Standard Deviation ($ \sigma $)**
   - **Description:** Measures the spread or dispersion of the distribution.
   - **Interpretation:** Determines the width of the bell curve; a larger $ \sigma $ results in a flatter and wider curve, while a smaller $ \sigma $ produces a steeper and narrower curve.

**Formula:**

The probability density function (PDF) of the Normal Distribution is given by:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{(x - \mu)^2}{2\sigma^2} }
$$

Where:
- $ x $ is the variable.
- $ \mu $ is the mean.
- $ \sigma $ is the standard deviation.
- $ e $ is the base of the natural logarithm.

### 🔹 **Properties**

- **Symmetry:** The distribution is perfectly symmetric around the mean $ \mu $.
- **Asymptotic:** The tails of the distribution approach, but never touch, the horizontal axis.
- **Empirical Rule:**
  - Approximately 68% of the data falls within $ \mu \pm \sigma $.
  - About 95% falls within $ \mu \pm 2\sigma $.
  - Nearly 99.7% lies within $ \mu \pm 3\sigma $.

### 📊 **Visualizing the Normal Distribution**

Visual representations aid in comprehending the characteristics of the Normal Distribution and assessing how closely the data follows this distribution.



```python
# Normal Distribution Visualization

from scipy.stats import norm

# Parameters
mu_sr = df['sr'].mean()
sigma_sr = df['sr'].std()

# Generate values for PDF
x = np.linspace(mu_sr - 4*sigma_sr, mu_sr + 4*sigma_sr, 1000)
pdf = norm.pdf(x, mu_sr, sigma_sr)

# Plot Histogram and PDF
plt.figure(figsize=(10,6))
sns.histplot(df['sr'], bins=30, kde=False, color='skyblue', stat='density', label='Strike Rate Histogram')
plt.plot(x, pdf, color='purple', label='Normal PDF')
plt.title('Normal Distribution of Sachin’s Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Density')
plt.legend()
plt.show()
```

**Explanation:**  
This plot overlays the **Normal PDF** on top of the histogram of Sachin's strike rates. The alignment of the bell curve with the histogram indicates how closely the strike rates follow a normal distribution.

### **Q-Q Plot (Quantile-Quantile Plot)**

```python
# Q-Q Plot to Assess Normality

from scipy.stats import probplot

plt.figure(figsize=(8,6))
probplot(df['sr'], dist="norm", plot=plt)
plt.title('Q-Q Plot of Sachin’s Strike Rate')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Strike Rate')
plt.show()
```

**Explanation:**  
A **Q-Q Plot** compares the quantiles of Sachin's strike rates with the quantiles of a standard normal distribution. If the points lie approximately along the reference line, it suggests that the data follows a normal distribution.

### 📈 **Interpreting the Normal Distribution**

Understanding the Normal Distribution allows for:

- **Standardization:** Converting data to a standard normal distribution (mean 0, standard deviation 1) using z-scores.
- **Probability Calculations:** Determining the likelihood of data falling within specific ranges.
- **Statistical Inference:** Many statistical tests assume normality, enabling the use of parametric methods.

**💡 Significance:**  
The Normal Distribution is pivotal due to the **Central Limit Theorem**, which states that the distribution of sample means approximates a normal distribution, regardless of the original data distribution, provided the sample size is sufficiently large.
