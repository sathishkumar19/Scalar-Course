```python
import pandas as pd

data = {
            'name': ['Mark', 'Ramilla', 'Deb', 'Laxman'],
            'profession': ['dev', 'mle', 'mle', 'hr'],
            'gender': ['male', 'female', 'male', 'male'],
            'age': [21, 20, 30, 27],
            'review': ['need improvement', 'Can be improved', 'Scope of improvement', 'Great'],
            'rating': [10, 5, 7, 9]
        }

df = pd.DataFrame(data)
df
```

```python
df_onlymale = df[df['gender'] =='male']
df_onlymale
```

```python
import pandas as pd
def specific_customers(df):
    ''' df is a dataframe with columns ['name', 'profession', 'gender', 'age', 'review', 'rating']
        Output -> A series with required names is expected to be returned'''
    # YOUR CODE GOES HERE
    
    # Select male professionals
    df_new = df[df['gender'] =='male']
    
    # Filter for age group 23-30
    df_new = df_new[(df_new['age'] >= 23) & (df_new['age'] <=30)]
    
    return df_new['name']
```
